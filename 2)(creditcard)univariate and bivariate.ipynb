```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("creditcard.csv")
dataset
```

```python
dataset=dataset.dropna(axis=0)
```

```python
dataset
```

```python
dataset.describe()
```

```python
import matplotlib.pyplot as plt

# Plotting histograms for numerical features
dataset.hist(bins=50, figsize=(20, 15))
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
import seaborn as sns

# Boxplots to check for outliers in numerical features
plt.figure(figsize=(12, 6))
sns.boxplot(data=dataset)
plt.xticks(rotation=90)
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each feature
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows where any value is outside the defined bounds
dataset_no_outliers = dataset[~((dataset < lower_bound) | (dataset > upper_bound)).any(axis=1)]

# Print the shape of the dataset before and after removing outliers
print(f"Original dataset shape: {dataset.shape}")
print(f"Dataset shape after removing outliers: {dataset_no_outliers.shape}")
```

```python


# Plot histograms for each feature to see the distribution
plt.figure(figsize=(15, 8))
dataset_no_outliers.hist(bins=50, figsize=(15, 8))
plt.tight_layout()
plt.show()
```

# Steps to Perform Bivariate Analysis on Credit Card Dataset:


```python
#1. Bivariate Analysis Between Features and the Target Variable (Class):1. Bivariate Analysis Between Features and the Target Variable (Class):


#1.1. Boxplots:

Boxplots are useful to visualize the distribution of continuous features for different categories of the target variable (Class = 1 for fraud, Class = 0 for non-fraud).
```

```python


# Box plot for a continuous feature against the target (fraud or non-fraud)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='V1', data=dataset)  # Example with 'V1' feature
plt.title('Boxplot of V1 by Class (Fraud vs Non-Fraud)')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('V1')
plt.show()
```

```python
# Box plot for a continuous feature against the target (fraud or non-fraud)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='V2', data=dataset)  # Example with 'V1' feature
plt.title('Boxplot of V1 by Class (Fraud vs Non-Fraud)')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('V2')
plt.show()
```

```python

# Violin plot to check the distribution of the features with respect to Class
plt.figure(figsize=(10, 6))
sns.violinplot(x='Class', y='V1', data=dataset)
plt.title('Violin Plot of V1 by Class (Fraud vs Non-Fraud)')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('V1')
plt.show()
```

Bivariate Analysis Between Features (Correlation Analysis):

# Calculate the correlation matrix for independent features
correlation_matrix = dataset.drop('Class', axis=1).corr()  # Drop target 'Class'

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()


```python

# Count plot to check the distribution of fraud (1) vs non-fraud (0) transactions
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=dataset)
plt.title('Distribution of Fraud vs Non-Fraud Transactions')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
# Example bar plot for a feature against target class
sns.barplot(x='V1', y='Class', data=dataset)
plt.title('Bar Plot of V1 by Class')
plt.xlabel('V1')
plt.ylabel('Class (Fraud vs Non-Fraud)')
plt.show()
```

#Summary of the 4 Plots:

```python
1)Boxplot: Shows the distribution of continuous variables across different categories of the target (Class), highlighting the median and detecting outliers.
2)Violin Plot: Offers a more detailed view of the distribution and density of features across the target classes, combining a boxplot and a density plot.
3)Correlation Heatmap: Displays the correlation between features to understand linear relationships. It helps identify redundant features.
```
